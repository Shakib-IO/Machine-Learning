```
#Importent libraries importing
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm   #importing the svm

from sklearn.datasets.samples_generator import make_blobs #make up random data

#we now create 40 separable points and we make 40 lines of data
#with 2 centers and random_state =20 means each group have 20 piece of different data
X ,y = make_blobs(n_samples=40 , centers=2 ,random_state = 20)


#Fit the model don't regularize for illustration purposes
clf = svm.SVC(kernel = 'linear' , C=1)
clf.fit(X,y)

#this from numpy array x[:,0].... 
#s(sze of dots)=30
#c(color)
plt.scatter(X[:,0], X[:, 1], c=y, s=50 , cmap=plt.cm.Paired )
plt.show()
```

```
#using to predict unknown data
newData = [[3,4],[5,6]]
print(clf.predict(newData))
```

```
#Now fit the model and don't regularize for illustration purposes

#clf=svm.SVC(kernel = 'linear', C=1000 )
#clf.fit(X,y)

#this from numpy array x[:,0].... 
#s(sze of dots)=30
#c(color)
plt.scatter(X[:,0], X[:, 1], c=y, s=30 , cmap=plt.cm.Paired )

#Plot the decision Fucntion
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()


#create grid to evaluate the model (x/y)lim=(x/y)limit
xx = np.linspace(xlim[0] , xlim[1] , 30)
yy = np.linspace(ylim[0] , ylim[1] , 30)

YY , XX = np.meshgrid(yy , xx)

xy = np.vstack([XX.ravel(), YY.ravel() ]).T

Z = clf.decision_function(xy).reshape(XX.shape)

#plot decision boundary and margin
ax.contour(XX , YY , Z , colors='k' , levels=[-1,0,1] , alpha =0.5 , 
           linestyles=['--', '-', '--'])


#plot support vectors
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[: , 1] , 
           s = 100 , linewidth=1 , facecolors='none')
plt.show()

```
